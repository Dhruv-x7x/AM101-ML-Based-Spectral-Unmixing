## Overview of the HYDICE Dataset

"The data set was acquired by imaging an area of Michigan, USA, with the Hyperspectral Digital Imagery Collection Experiment (HYDICE) sensor. With a resolution of approximately 2 m, the original image contains 210 bands from the visible to the near-infrared spectrum. After removing the low signal-to-noise ratio (SNR) and water absorption bands, 175 bands are reserved for experiments. An image containing 80Ã—100 pixels is cropped for experiments." sxt1996 from GitHub [https://github.com/sxt1996/HYDICE]

Initial experiments will be performed on this image before moving on to potentially larger datasets with much higher spatial and spectral resolution. The goal of this notebook is to understand the structure of the data cube and explore very basic unmixing algorithms.
